## 10) Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

```python
battingDf_24 = battingDf.copy()
```

```python
battingDf_24['OBP'] = (battingDf_24['H'] + battingDf_24['BB'] + battingDf_24['IBB'] + battingDf_24['SH'] + battingDf_24['SF'])/battingDf_24['AB']
```

```python
battingDf_24['PA'] = battingDf_24['AB'] + battingDf_24['BB'] + battingDf_24['IBB'] + battingDf_24['SH'] + battingDf_24['SF']
```

```python
battingDf_24['AVG'] = battingDf_24['H']/battingDf_24['AB']
```

```python
battingDf_25 = battingDf_24[(battingDf_24['PA']>=500) & (battingDf_24['yearID']==1977) ]
```

```python
battingDf_25.sort_values('OBP',ascending = 0)[:5]
```
